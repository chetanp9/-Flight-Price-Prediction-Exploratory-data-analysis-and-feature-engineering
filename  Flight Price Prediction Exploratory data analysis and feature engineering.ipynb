## feature engineering

1. Date_of_Journey is an object so we have to convert it into datetime dtype or derived feature

```python

final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
final_df.head()
```

```python
## converting date months and year into int
final_df['Date']=final_df['Date'].astype(int)
```

```python
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python

final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(10)
```

```python
##2.arrival time
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
final_df.head(1)
```

```python
## derived hour and min arival time
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min']=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
## converting the hour and min into int
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
#droping the arival time
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
## for departure time
final_df['Dep_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_min']=final_df['Dep_Time'].str.split(':').str[1]
##converting into int
final_df['Dep_hour']=final_df['Dep_hour'].astype(int)
final_df['Dep_min']=final_df['Dep_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
## now for Total_Stops
final_df['Total_Stops'].unique()
```

```python
##Total_Stops having null values
final_df[final_df['Total_Stops'].isnull()]
```

```python
## maping stops to a int value using map or target ordinal encoding
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'2 stops':2,'1 stop':1,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df.head(1)
```

```python
## droping the route feature as not neede we have total stops to calculate price
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
## now for durations features
final_df.info()
```

```python
final_df[final_df['duration_hour']=='5m']
```

```python
##droping the records with 5m
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
## deriving the hour from durations and converting into minutes

final_df['duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df['duration_hour']=final_df['duration_hour'].astype(int)
```

```python
## converting the durations hour into min
final_df['duration_hour']=final_df['duration_hour']*60
```

```python
final_df['durations_min']=final_df['Duration'].str.split('m').str[0].str.split(' ').str[1]
```

```python
##replacing the null value in durattions_min with 0
final_df['durations_min']=final_df['durations_min'].fillna(value=0)
```

```python

final_df['durations_min']=final_df['durations_min'].astype(int)
```

```python
final_df.head()
```

```python
## adding the duration hour and minutes and saving into final durations
final_df['final_durations_min']=final_df['duration_hour']+final_df['durations_min']
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
## droping the durationd hour ,minutes and Durations as not need we have final durations
final_df.drop('duration_hour',axis=1,inplace=True)
final_df.drop('durations_min',axis=1,inplace=True)
final_df.drop('Duration',axis=1,inplace=True)
final_df.drop('final_durations',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```
